```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
import warnings

warnings.filterwarnings('ignore')

iris = sb.load_dataset("iris")

iris.info()

sb.set_style(style='whitegrid')
```

```python
sb.violinplot(x="species", y="petal_length", data=iris, hue='species')
```

```python
sb.swarmplot(x="species", y='petal_length', data=iris, hue='species')
```

```python
sb.boxplot(x="species", y='petal_length', data=iris, hue='species')
```

```python
numeric_data = iris.select_dtypes(include=[float])

corr = numeric_data.corr()

sb.heatmap(corr, fmt='.3f', cmap="coolwarm", annot=True)

plt.show()
```
