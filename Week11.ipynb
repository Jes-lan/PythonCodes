```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
import warnings

warnings.filterwarnings('ignore')


data = pd.read_csv('Pokemon.csv', index_col=0)


sb.lmplot(x='Attack', y='Defense', data=data)
```

```python
data.head()
```

```python
sb.lmplot(x='Attack', y='Defense', fit_reg=False, hue="Stage", data=data)
```

```python
sb.lmplot(x='Attack', y='Defense', fit_reg=False, hue="Stage", data=data)

plt.ylim(0, None)
plt.xlim(0, None)
```

```python
sb.boxplot(data=data)
plt.xticks(rotation=-45)
```

```python
stats_data = data.drop(["Total", "Stage", "Legendary"], axis=1)

sb.boxplot(data=stats_data,
            notch=True,
            linewidth=2.5,
            width=0.5)
```

```python
sb.set_style("whitegrid")

sb.violinplot(x="Type 1", y="Attack", data=data)
plt.xticks(rotation=-45)
```

```python
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                    ]

sb.violinplot(x='Type 1', y='Attack', data=data, palette=pkmn_type_colors)

plt.xticks(rotation=-45)
```

```python
sb.swarmplot(x='Type 1', y='Attack', data=data, palette=pkmn_type_colors)

plt.xticks(rotation=-45)
```

```python
sb.violinplot(x='Type 1', y='Attack', data=data, inner=None, palette=pkmn_type_colors)

sb.swarmplot(x='Type 1', y='Attack', data=data, color='k', alpha=0.7)

plt.title('Attack by Type')
plt.xticks(rotation=-45)
```
